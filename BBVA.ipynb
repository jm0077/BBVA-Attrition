```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from adspy_shared_utilities import plot_class_regions_for_classifier_subplot
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
train_clientes = pd.read_excel('train_clientes.xlsx')
print(train_clientes.shape)
train_clientes.head()
```

```python
X = train_clientes.drop(['ATTRITION'], axis=1)
X.set_index('ID_CORRELATIVO',inplace=True)
print(X.shape)
X.head()
```

```python
y = train_clientes['ATTRITION']
print(y.shape)
y.head()
```

```python
train_req = pd.read_excel('train_requerimientos.xlsx')
train_req.set_index('ID_CORRELATIVO', inplace=True)
print(train_req.shape)
train_req.head()
```

```python
train_req_encoded.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_cleaned, y, random_state = 0)


#fig, subaxes = plt.subplots(1, 1, figsize=(6, 6))

clf_RF = RandomForestClassifier().fit(X_train, y_train)
#title = 'Random Forest Classifier, default settings'
#plot_class_regions_for_classifier_subplot(clf_RF, X_train, y_train, X_test, y_test, title, subaxes)
#plt.show()

train_score = clf_RF.predict(X_train)
f1_train = f1_score(y_train, train_score, average='micro')  
print("Train F1 score: %f" % f1_train)
prediction = clf_RF.predict(X_test)
f1_test = f1_score(y_test, prediction, average='micro')  
print("Test F1 score: %f" % f1_test)
```
